5.count survival by gender, categories wise age of various classes. 

```python
 titanic_data['category']  = pd.cut(titanic_data['age'],[0,10,30,60,80])
 titanic_data.head()
```

```python
titanic_data.pivot_table('survived',index=['sex','category'],columns='pclass', aggfunc='count').fillna(0)
```

Survival count by gender

```python
titanic_data.groupby(['survived','sex']).age.count().reset_index().pivot(columns='survived',index='sex',values='age').reset_index()
```

number of women and men were in a particular cabin class.

```python
table = titanic_data.groupby(['who','pclass']).age.count().reset_index().pivot(columns='who',index='pclass',values='age')
table
```
